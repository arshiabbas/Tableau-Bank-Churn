```python
bank_numerical = bank.select_dtypes(exclude='object')
bank_numerical.columns
```

```python
sns.boxplot(data=bank_numerical, y='CLIENTNUM')
```

```python
bank_numerical_1 = bank_numerical[['Customer_Age', 'Dependent_count', 'Months_on_book',
       'Total_Relationship_Count', 'Months_Inactive_12_mon']]
bank_numerical_2 = bank_numerical[['Contacts_Count_12_mon', 'Credit_Limit', 'Total_Revolving_Bal',
       'Avg_Open_To_Buy', 'Total_Trans_Amt', 'Total_Trans_Ct',
       'Avg_Utilization_Ratio']]
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(data=bank_numerical_1)
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(data=bank_numerical_2)
```

```python
# removing outliers with z_score
def remove_outliers_z(bank, column):
    z_scores = (bank[column] - bank[column].mean())/bank[column].std()
    bank = bank[(z_scores < 3) & (z_scores > -3)]
    return bank

for column in bank_numerical:
    bank = remove_outliers_z(bank, column)
```

```python
bank_numerical_aftr_out = bank.select_dtypes(exclude='object')
bank_numerical_aftr_out1 = bank_numerical_aftr_out[['Customer_Age', 'Dependent_count', 'Months_on_book',
       'Total_Relationship_Count', 'Months_Inactive_12_mon', 'Contacts_Count_12_mon']]
bank_numerical_aftr_out2 = bank_numerical_aftr_out[['Credit_Limit', 'Total_Revolving_Bal',
       'Avg_Open_To_Buy', 'Total_Trans_Amt', 'Total_Trans_Ct',
       'Avg_Utilization_Ratio']]
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(data=bank_numerical_aftr_out2)
```

```python
# removing outliers with z_score for 2nd time
def remove_outliers_z2(bank, column):
    z_scores = (bank[column] - bank[column].mean())/bank[column].std()
    bank = bank[(z_scores < 3) & (z_scores > -3)]
    return bank

for column in bank_numerical_aftr_out:
    bank = remove_outliers_z2(bank, column)
```
